This notebook is to calculate the cumulative sum based on month

```python
# libraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("monthly_non_growing.csv")
```

```python
df.drop(columns=['cumulative_sum','growing'], inplace=True)
```

```python
df['month'] = df['Year/Month'].apply(lambda x: x[-2:])
df['month'] = pd.to_numeric(df['month'], downcast="integer")
# df['month_t1'] = df['month'].shift(1)
# # replacing nan for the first row 
# df['month_t1'][0] = df['month'][0]
# df['month_t1'] = df['month_t1'].astype(int)
df
```

# !!! IMPORTANT
## Cumulative sum are based on index so make sure to reset the index properly before proceeding the below codes

```python
# tagging when it is November
df['tag'] = 0
df.loc[df['month']==11,'tag'] = 1
# identifying the index when there is a change in months
index_list = df.loc[df['tag']==1].index
index_list
```

```python
# calculating the cumulative sum
# first set of cumulative sum is calculated based on indexing
cum_sum_df = df[['mm']][:index_list[0]].cumsum()
# the rest of the cumulative sum are appended to the dataframe
for i in range(len(index_list)):
    if i != len(index_list)-1:
        cum_sum_df = cum_sum_df.append(df[['mm']][index_list[i]:index_list[i+1]].cumsum())
    else:
        # this is for the final set of cumulative sum
        cum_sum_df = cum_sum_df.append(df[['mm']][index_list[i]:].cumsum())
```

```python
# renaming the column name
cum_sum_df.rename(columns = {'mm': 'cumulative_sum'}, inplace=True)
```

```python
df
```

```python
# dropping unnecessary columns
# df.drop(columns=['month','tag'], inplace=True)
# merging the cumulative sum based on the index
df = df.merge(cum_sum_df, how='left', left_index=True, right_index=True)
df
```

```python
df = df.set_index('Year/Month')
```
