## the data limit is 32000, therefore csv file will be used for full data

```python
csv_file = "../../Database/NSW_Covid_cases/confirmed_cases_table4_location_likely_source.csv"
df_csv = pd.read_csv(csv_file)
```

```python
df_csv['daily_total'] = 1
df_csv
```

```python
df_daily = df_csv.groupby('notification_date')[['daily_total']].count()
df_daily.index = pd.to_datetime(df_daily.index)
df_daily
```

```python
import pandas as pd
from datetime import datetime

pd.date_range(start="2020-01-25",end="2021-10-10")
```

```python
df_new = pd.DataFrame(pd.date_range(start="2020-01-25",end="2021-10-10"), columns=['notification_date'])
df_new = df_new.set_index('notification_date')
df_new
```

```python
df_new = df_new.merge(df_daily, how='left', left_index=True, right_index=True)
df_new
```

```python
# replacing the missing values to zero
df_new.fillna(0, inplace=True)
```

## lag and lead

```python
#lag
df_new['t-1'] = df_new['daily_total'].shift(1)
df_new['t-2'] = df_new['daily_total'].shift(2)
df_new['t-3'] = df_new['daily_total'].shift(3)
#lead
df_new['t+1'] = df_new['daily_total'].shift(-1)
df_new['t+2'] = df_new['daily_total'].shift(-2)
df_new['t+3'] = df_new['daily_total'].shift(-3)
```

```python
df_new
```

# Moving average

```python
rolling_ma_3 = df_new['daily_total'].rolling(window=3).mean()
rolling_ma_3
```

```python
df_new['ma_3'] = df_new['daily_total'].rolling(window=3).mean()
df_new['ma_5'] = df_new['daily_total'].rolling(window=5).mean()
df_new['ma_7'] = df_new['daily_total'].rolling(window=7).mean()
df_new['ma_10'] = df_new['daily_total'].rolling(window=10).mean()
df_new['ma_14'] = df_new['daily_total'].rolling(window=14).mean()
```

```python
df_new
```

```python
df_new['ma_3'][:100].plot()
```

# Autocorrelation

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(df_new['daily_total'])
```
